```python
import cv2
import matplotlib.pyplot as plt         # Importing Image
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
img=cv2.imread('pushkar.jpg')
```

```python
img                                   # matrix or coordinate of Image
```

```python
cv2.imshow('pushkar',img)
plt.imshow(img)
plt.title('Original Image')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_rgb)
plt.title('Original image to gray_Scale')         #converted image in to gray scale
```

```python
inverted_img=cv2.bitwise_not(img_rgb)
plt.imshow(inverted_img)
plt.title("inverted image")              #inverted image
```

```python
inverted_img
```

```python
Blur_img=cv2.GaussianBlur(inverted_img, (21, 21), 0)
plt.imshow(Blur_img)
plt.title("Blur image")        #Blur image
```

```python
inverted_blur=cv2.bitwise_not(Blur_img)
plt.imshow(inverted_blur)                      #inverted blur Image
```

```python

pencil_sketch = cv2.divide(inverted_img, inverted_blur, scale=256.0)
plt.imshow(pencil_sketch)
plt.title("Pencile Sketch Image")
```

```python
cv2.imwrite("sketch.png", pencil_sketch )     #pencil sketch saved on dekstop
```
